```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to load
schools_data_to_load = r"schools_complete.csv"
students_data_to_load = r"students_complete.csv"
```

```python
# Read the student data file and store into a Pandas DataFrame
school_data_df = pd.read_csv(schools_data_to_load)
student_data_df = pd.read_csv(students_data_to_load)

#Cleaning Student Names and Replacing Substrings in a Pythong String

# Add each prefix and suffix to remove to a list
prefixes_suffixes = ["Dr.", "Mr.", "Ms.", "Mrs.", "Miss", "MD", "DDS", "DVM", "PHD"]

# Iterate through the words in the "prefixes_suffixes" List and replace them with an empty space,
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"grade")
    
# Check names.
student_data_df.head(10)
```

```python
#creates grade level average reading scores for each school
tenth_reading = school_data_complete_df.loc[school_data_complete_df['grade'] == '10th'].groupby('school_name')["reading_score"].mean()
eleventh_reading = school_data_complete_df.loc[school_data_complete_df['grade'] == '11th'].groupby('school_name')["reading_score"].mean()
twelfth_reading = school_data_complete_df.loc[school_data_complete_df['grade'] == '12th'].groupby('school_name')["reading_score"].mean()

#merges the reading score averages by school and grade together
reading_scores = pd.DataFrame({
        "10th": tenth_reading,
        "11th": eleventh_reading,
        "12th": twelfth_reading
})
reading_scores = reading_scores[['10th', '11th', '12th']]
reading_scores.index.name = "school_namel"

#format
reading_scores.style.format({ "10th": '{:.1f}', 
                             "11th": "{:.1f}", 
                             "12th": "{:.1f}"})
```

```python
#creates grade level average math scores for each school 
tenth_math = school_data_complete_df.loc[school_data_complete_df['grade'] == '10th'].groupby('school_name')["math_score"].mean()
eleventh_math = school_data_complete_df.loc[school_data_complete_df['grade'] == '11th'].groupby('school_name')["math_score"].mean()
twelfth_math = school_data_complete_df.loc[school_data_complete_df['grade'] == '12th'].groupby('school_name')["math_score"].mean()
budget = school_data_complete_df.loc[school_data_complete_df['grade'] == 'budget'].groupby('school_name') ["math_score"].count()

math_scores = pd.DataFrame({
        "10th": tenth_math,
        "11th": eleventh_math,
        "12th": twelfth_math
})
math_scores = math_scores[['10th', '11th', '12th']]
math_scores.index.name = "school_name"

#show and format
math_scores.style.format({"10th": '{:.1f}', 
                          "11th": "{:.1f}", 
                          "12th": "{:.1f}", 
                       })
```
